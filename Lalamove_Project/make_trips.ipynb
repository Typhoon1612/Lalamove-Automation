```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
num_trips = 100
start_date = datetime.now() - timedelta(days=30)

data = {
    'Trip_ID':[f'LALA-{i:04d}' for i in range(1, num_trips + 1)],
    'Date':[start_date + timedelta(days=np.random.randint(0,30)) for _ in range(num_trips)],
    'Distance_KM':np.round(np.random.uniform(5.0, 45.0, num_trips), 2),
    'Lalamove_Earnings_RM':np.round(np.random.uniform(10.0, 80.0, num_trips), 2)

}
```

```python
df = pd.DataFrame(data)
df.to_csv('lalamove_trips_march.csv', index=False)
print("Fake dataset 'lalamove_trips_march.csv' created successfully.")
```

```python
# 1. Create 50 new trips for April
num_april_trips = 50
april_start = datetime.now()

april_data = {
    'Trip_ID': [f'LALA-APR-{i:04d}' for i in range(1, num_april_trips + 1)],
    'Date': [april_start + timedelta(days=np.random.randint(0, 30)) for _ in range(num_april_trips)],
    'Distance_KM': np.round(np.random.uniform(5.0, 45.0, num_april_trips), 2),
    'Lalamove_Earnings_RM': np.round(np.random.uniform(10.0, 80.0, num_april_trips), 2)
}

# 2. Save this as a NEW file in the SAME folder
df_april = pd.DataFrame(april_data)
df_april.to_csv('lalamove_trips_april.csv', index=False)
print("April data created! Check your folder.")
```
